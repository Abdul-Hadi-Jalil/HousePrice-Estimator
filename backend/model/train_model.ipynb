```python
import pandas as pd

df = pd.read_csv(r"..\dataset\Cleaned_data_for_model.csv")

del df["purpose"]
del df['Unnamed: 0']

df
```

```python
pd.isnull(df)
```

```python
cols_to_fill = ['price', 'baths', 'bedrooms', 'Area_in_Marla']
df[cols_to_fill] = df[cols_to_fill].fillna(df[cols_to_fill].mean())
```

```python
cols_to_fill_with_mod = ['property_type', 'location', 'city']
df[cols_to_fill_with_mod] = df[cols_to_fill_with_mod].fillna(df[cols_to_fill_with_mod].mode().iloc[0])
```

```python
df['property_type'] = df['property_type'].map({'Flat':0, "House": 1})
```

```python
from sklearn.preprocessing import TargetEncoder

encoder = TargetEncoder()

# Correct: Pass as DataFrame (2D) using double brackets
df['location_encoded'] = encoder.fit_transform(df[['location']], df['price'])  # Note [[ ]]
df['city_encoded'] = encoder.fit_transform(df[['city']], df['price'])          # Note [[ ]]
```
